## Train-Test Splitting

```python
import numpy as np
#For learning purpose 
def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
```

```python
#print(f"Rows in train set: {len(train_set)}\nRows in test set: {len(test_set)}")
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
print(f"Rows in train set: {len(train_set)}\nRows in test set: {len(test_set)}")
```

## Scikit-learn Design

Primarily, three types of objects
1. Estimators - It estiates some paramter based on some datase. Ex. imputer. 
                It has a fir method and a transform method.Fit Method - Fits the dataset and calculates the internal parameters

2. Transformers - Transform method takes input and returns output based on the learnings from fit().
                  It also has a convenience function called fit_transform() which fits and then transform.
                  
3. Predictors  - LinearRegressionModel is the example. Fit() and Predict() are two common functions.It also gives 
                 score() function which will evaluate the predictions.

## Feature Scaling

Priarily, two types,
1. Min-Max scaling (Normalization)
       (value - min) / (max - min)
       sklearn provides a class called MinMaxScaler for this

2. Standardization
        (value-mean) / std
        sklearn provides a class called StandardScaler for this

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
#   ...... add as many as you want is your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing_tr)
```

```python
housing_num_tr
```

```python
housing_num_tr.shape
```

## Selecting a desired model for Dragon Real Estates

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
# model = LinearRegression()
# model = DecisionTreeRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
some_data = housing.iloc[:5]
```

```python
some_labels = housing_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluating the model

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels, housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

## Using better evaluation technique - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_labels, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standard Deviation: ", scores.std())
```

```python
print_scores(rmse_scores)
```

## Saving the model

```python
from joblib import dump, load
dump(model, 'Dragon.joblib')
```

## Testing the Model on test data

```python
X_test = strat_test_set.drop("MEDV", axis=1)
Y_test = strat_test_set["MEDV"].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
# print(final_predictions, list(Y_test))
```
